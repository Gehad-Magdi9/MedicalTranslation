```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
from google.colab import drive
drive.mount('/content/drive')
# تحميل النموذج المدرب
model_path = "/content/drive/My Drive/medical-translation-model/test2"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
```

```python
#will be connected with GUI
import torch

# Translate Function
def translate(text):
    if not text.strip():
        return "النص المدخل فارغ. يرجى إدخال نص للترجمة."
    # Prepare the input
    inputs = tokenizer(text, return_tensors="pt", max_length=128, truncation=True, padding="max_length")
    # Generate The Translation
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128, num_beams=5, early_stopping=True)
    # Decode The Translation
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text
```

```python
import gradio as gr

# Gradio Interface
interface = gr.Interface(
    fn=translate,            # Translate function
    inputs="text",
    outputs="text",
    title="Medical Translator",
    description="ترجمة النصوص الطبية من الإنجليزية إلى العربية باستخدام نموذج مدرب."
)

interface.launch(share=True)
```
